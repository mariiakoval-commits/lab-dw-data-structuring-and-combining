## Challenge 1: Combining & Cleaning Data

In this challenge, we will be working with the customer data from an insurance company, as we did in the two previous labs. The data can be found here:

But this time, we got new data, which can be found in the following 2 CSV files located at the links below.

Note that you'll need to clean and format the new data.

Observation:

- Another option would be to read the clean file you saved in the previous lab, and just clean the two new files and concatenate the three clean datasets

```python
# Your code goes here

import pandas as pd
import numpy as np


data_client1 = pd.read_csv('cleaned_customer_data.csv')
data_client2 = pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv')
data_client3 = pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv')

data_client2 
data_client2.columns = data_client2.columns.str.lower()
data_client2 
```

```python
data_client3
data_client3.columns = data_client3.columns.str.lower()
```

```python
data_client2.columns = data_client2.columns.str.replace(' ','_')
data_client3.columns = data_client3.columns.str.replace(' ','_')
data_client2 = data_client2.rename(columns = {'st': 'state'})
data_client3 = data_client3.rename(columns = {'st': 'state'})
```

```python
display(data_client1) 
display(data_client2)
display(data_client3)
```

```python
data_client1.dtypes
```

```python
data_client2.dtypes
```

```python
data_client3.dtypes
```

```python

data_client2['customer_lifetime_value'] = data_client2['customer_lifetime_value'].str.replace('%', '')
data_client2['customer_lifetime_value'] = data_client2['customer_lifetime_value'].astype(float)
data_client3['customer_lifetime_value'] = data_client3['customer_lifetime_value'].astype(float)
data_client3['customer_lifetime_value'] = data_client3['customer_lifetime_value'].round(2)
```

```python
gender_mapping = {
    'Male': 'M', 'male': 'M',
    'female': 'F', 'Female': 'F',
    'F': 'F', 'M': 'M'
}
data_client2['gender'] = data_client2['gender'].map(gender_mapping)

data_client2['gender'].unique()
```

```python
data_client1['monthly_premium_auto'] = data_client1['monthly_premium_auto'].astype(int)
```

```python
def get_complains(value):
    if type(value)==str:
        return value.split('/')[1]

data_client1['number_of_open_complaints'] = data_client1['number_of_open_complaints'].apply(get_complains)
data_client2['number_of_open_complaints'] = data_client2['number_of_open_complaints'].apply(get_complains)
```

```python

data_client2['vehicle_class'] = data_client2['vehicle_class'].str.replace('Sports Car', 'Luxury')
data_client2['vehicle_class'] = data_client2['vehicle_class'].str.replace('Luxury SUV', 'Luxury')
data_client2['vehicle_class'] = data_client2['vehicle_class'].str.replace('Luxury Car', 'Luxury')

data_client3['vehicle_class'] = data_client3['vehicle_class'].str.replace('Sports Car', 'Luxury')
data_client3['vehicle_class'] = data_client3['vehicle_class'].str.replace('Luxury SUV', 'Luxury')
data_client3['vehicle_class'] = data_client3['vehicle_class'].str.replace('Luxury Car', 'Luxury')
```

```python
data_client3.isnull()
```

```python
data_client_full = pd.concat([data_client1, data_client2, data_client3], ignore_index=True)
data_client_full. head(50)
```

```python
data_client_full.duplicated(subset=['customer', 'customer_lifetime_value']).sum()
```

```python
data_client_full.duplicated().sum()
```

```python
data_client_full.drop_duplicates(keep='last', inplace=True)
```

```python
# Your code goes here

import pandas as pd
import numpy as np

marketing_customer_analysis = pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv')


marketing_customer_analysis = marketing_customer_analysis.drop(['unnamed:_0'], axis=1)
marketing_customer_analysis
```

```python
revenue_by_channel = pd.pivot_table(marketing_customer_analysis, values='total_claim_amount',  index='sales_channel', 
    aggfunc='sum'
)

revenue_by_channel = revenue_by_channel.round(2)
revenue_by_channel
```

```python
customer_sex_edd = pd.pivot_table(marketing_customer_analysis, index = 'gender', 
                                  columns = 'education', 
                                  values = 'customer_lifetime_value', aggfunc = 'mean' )

customer_sex_edd 
```
